```python
words=["earing","eats","eaten","writing","writes","programming","programs","history","finally"]
```

```python
##1st technique
##porterstemmer
words=["earing","eats","eaten","writing","writes","programming","programs","history","finally"]

from nltk.stem import PorterStemmer
stemming=PorterStemmer()
```

```python
for word in words:
    print(word+"--->"+stemming.stem(word))
```

### disadvantage of stemming history--->histori not give proper result for some example 


```python
##example
stemming.stem("congratulations")
```

### regexpstemmer class->NLTK has regexpstemmer class with the help
of which we can easily implement regular expression stemmer algo  

```python
from nltk.stem import RegexpStemmer

reg_stemmer=RegexpStemmer('ing|s$|able$',min=4)
## ing means where where ing remove ,able$ at end of any word able remove 

reg_stemmer.stem("easing")
```

```python
reg_stemmer.stem("ingeasing")
reg_stemmer.stem("unable")
```

### sonwball stemmmer 



```python
from nltk.stem import SnowballStemmer

snowballstemmer=SnowballStemmer('english')

for word in words:
    print(word+"-->"+snowballstemmer.stem(word))
```

```python
stemming.stem("fairly"),stemming.stem("sportingly")
```

```python
snowballstemmer.stem("fairly"),snowballstemmer.stem("sportingly")
```
