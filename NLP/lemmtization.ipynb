## lemmitization

technique is like stemming output called lemma the output of stemming after
lemmitization we will get valid word means same thing 

main give exact root word 

```python
from nltk.stem import WordNetLemmatizer

lemmatizer=WordNetLemmatizer()

```

```python
''' 
pos-noun-n
verb -v
adjective-a
adverb-r 
'''

lemmatizer.lemmatize("goes",pos='v')
```

```python
words=["visionary","eats","eaten","writing","writes","programming","programs","history","finally"]
for word in words:
    print(word+"--->"+lemmatizer.lemmatize(word,pos='v'))
```
