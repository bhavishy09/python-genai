```python
students=[("aman",101,80,90),
           ("Manya", 102, 78, 82),
    ("Madhav", 103, 92, 88),
    ("Manav", 104, 65, 70),
    ("Apeksha", 105, 90, 95),
    ("Arindham", 106, 92, 88)  
          ]



##nested dict 
student_dict= {}

for name,id,mid,end in students:
    avg=round((mid+end)/2,2)
    student_dict[id]={
        "name":name,
        "marks":(mid,end),
        "avg":avg
    }


print(student_dict)


for id,mid,end in students:
    classavg=(mid+end)

  
 
```

```python
classavg=round(sum(d["avg"] for d in student_dict.values()) / len(student_dict),2)

print(classavg)
```

```python
above_avg=[(d["name"],d["avg"]) for d in student_dict.values() if d["avg"]>classavg
           ]

avg_sort=sorted(above_avg)

for name,avg  in avg_sort:
      print(f"{name} -> {avg}")
```
