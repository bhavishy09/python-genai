```python
def count_vowel(s):
    
    vowels='aeiou'
    s=s.lower()
    count=0

    for ch in s:
        if ch in vowels:
            count+=1
    return count 

s="abhishkek"
ko=count_vowel(s)
print(ko)
```

```python
def longest_unique_substring(s):
    longest = ""   # store longest substring
    n = len(s)

    # check all substrings
    for i in range(n):
        seen = set()   # store visited chars
        current = ""
        for j in range(i, n):
            if s[j] in seen:
                break  # repeat mila to break
            seen.add(s[j])
            current += s[j]
            if len(current) > len(longest):
                longest = current
    return longest, len(longest)


# Example
print(longest_unique_substring("abcabcbb"))  # ('abc', 3)
print(longest_unique_substring("pwwkew"))    # ('wke', 3)
print(longest_unique_substring("bbbbb"))     # ('b', 1)
```

```python
def is_palindorome(s):
    i=0
    j=len(s)-1
    while i<j:
     if s[i] !=s[j]:
        return False
    i+=1
    j-=1
    return True

s="abccba"

print(is_palindorome(s))
```

```python
def factorial(n):
    if n <= 1: return 1
    return n * factorial(n-1)
```

```python
def reverse(s):
    letter=[ch for ch in s if ch.isalpha()]
    results=[]
    for ch in s:
        if ch.isalpha():
            results.append(letter.pop())
        else:
            results.append(ch)
        
    return "".join(results)

s="a-vk-klo-lo"
print(reverse(s))
```

```python
from collections import Counter

def first_unique_char(s):
    count = Counter(s)
    for ch in s:
        if count[ch] == 1:
            return ch
    return None

print(first_unique_char("leetcode"))  # l
print(first_unique_char("aabb"))      # None
```

```python
def compress_string(s):
    result = ""
    count = 1
    for i in range(1, len(s)+1):
        if i < len(s) and s[i] == s[i-1]:
            count += 1
        else:
            result += s[i-1] + str(count)
            count = 1
    return result

print(compress_string("aaabbc"))  # a3b2c1
```

```python
import numpy as np

def is_prime(n):
    if n<2:
        return False
    for i in range(2,int(n**0.5)+1):
        if n%i==0:
            return False
        return True
    
def ques2(A,B):

    for i in range(min(A.shape)):
        if A[i,i]%3!=0:
            A[i,i]=-A[i,i]
    print(A)

    min_val=np.min(B)
    for i in range(min(B.shape)):
        if is_prime(B[i,i]):
            B[i,i]=min_val
    print(B)
   
    if A.shape[1]==B.shape[0]:
        m=A @ B
        print(m)
        print("is symmetric?",np.array_equal(m,m.T))


A = np.array([[3,2,1],[4,5,6],[7,8,9]])
B = np.array([[2,3,5],[6,7,11],[13,17,19]])
ques2(A, B)
```
