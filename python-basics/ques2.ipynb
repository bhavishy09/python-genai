```python
goods=[
    (301 ,"Python Basics", "John Doe", 120, 110),
    (302, "Data Science Essentials" ,"Alice Smith" ,90 ,85),
    (303, "Machine Learning 101", "Bob Johnson ",150 ,148),
    (304, "AI Revolution"," Carol Lee ",200 ,195),
    (305, "Deep Learning Simplified", "Dave Clark", 75, 70)

]

books={}

for book_id,title,auth,iss,returned in goods:
    return_rate=round((returned/iss)*100,2)
    books[book_id]={
          "title": title,
        "author": auth,
        "issued": iss,
        "returned": returned,
        "return_rate": return_rate

    }

print(books)
```

```python


high_rate=max(book["return_rate"] for book in books.values())

print(high_rate)
```

```python
top_books = [
    (book_id, details)
    for book_id, details in books.items()
    if high_rate - 5 <= details["return_rate"] < high_rate
]

print(top_books)
```

```python
boooks={}

for book_id,title,auth,iss,returned in goods:
    if iss > 150:
        boooks[title]="High Circulation"
    elif 100 <= iss <= 150:
         boooks[title]="Moderate Circulation"
    else :
         boooks[title]= "Low Circulation"



print(boooks)
    
```
