```python
## read a file 

with open('open.txt','r') as file:
    content=file.read()
    print(content)

with open('open.txt','w') as file:
    file.write("jai shree ram")


with open('open.txt','w+') as file:
    file.write("jai shree ram ji")
    content=file.read()
    print(content)


with open('open.txt','a+') as file:
    file.write("hare krishna")
    content=file.read()
    print(content)

    


   



```
