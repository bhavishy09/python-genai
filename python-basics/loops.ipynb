```python
## Loop Control Statements

## break
## The break statement exits the loop permaturely

## break sstatement

for i in range(10):
    if i==5:
        break
    print(i)
   
```

```python
for i in range(3):
    for j in range(2):
        print(f"i:{i} and j:{j}")
```

```python
sum=0
N=10
count=1

while count<=N:
    sum=count+sum
    count=count+1

    

print("Sum of first 10 natural number:",sum)
```

```python
## Example- Prime numbers between 1 and 100

for num in range(1,101):
    if num>1:
        for i in range(2,num):
            if num%i==0:
                break
        else:
            print(num)
         
       
                
```
