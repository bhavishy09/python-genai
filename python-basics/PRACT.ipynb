```python
import matplotlib.pyplot as plt

# Data from the question
weeks = [1, 2, 3, 4, 5]
refrigerator = [45, 48, 50, 47, 49]
washing_machine = [25, 28, 26, 30, 29]
air_conditioner = [70, 75, 80, 76, 78]

# Calculate total energy consumption per week
total_energy = [r + w + a for r, w, a in zip(refrigerator, washing_machine, air_conditioner)]

# Find the week with the highest total energy consumption
max_week = weeks[total_energy.index(max(total_energy))]
max_value = max(total_energy)

# Plotting the line chart
plt.figure(figsize=(8, 5))
plt.plot(weeks, refrigerator, marker='o', color='blue', label='Refrigerator')
plt.plot(weeks, washing_machine, marker='s', color='green', label='Washing Machine')
plt.plot(weeks, air_conditioner, marker='^', color='red', label='Air Conditioner')

# Annotate the week with the highest total energy consumption
plt.annotate(f'Highest total: {max_value} kWh (Week {max_week})',
             xy=(max_week, max_value / 3), xytext=(max_week, max_value / 3 + 10),
             arrowprops=dict(facecolor='black', shrink=0.05),
             fontsize=9, bbox=dict(facecolor='yellow', alpha=0.3))

# Add labels, title, and legend
plt.xlabel('Week')
plt.ylabel('Energy Consumption (kWh)')
plt.title('Weekly Energy Consumption Trend')
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.show()
```

```python
# -------------------------------------------
# Import Required Libraries
# -------------------------------------------
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt

# -------------------------------------------
# Step 1: Read and Parse the HTML File
# -------------------------------------------
with open('product_list.html', 'r', encoding='utf-8') as file:
    html_content = file.read()

soup = BeautifulSoup(html_content, 'html.parser')

# -------------------------------------------
# Step 2: Extract Data from Table
# -------------------------------------------
rows = soup.find_all('tr')
data = []

for row in rows:
    cols = row.find_all('td')
    if len(cols) == 4:  # only valid rows
        sno = cols[0].text.strip()
        title = cols[1].text.strip()
        category = cols[2].text.strip()
        price = cols[3].text.strip().replace('$', '')
        data.append([sno, title, category, float(price)])

# -------------------------------------------
# Step 3: Create a Pandas DataFrame
# -------------------------------------------
df = pd.DataFrame(data, columns=['S.No', 'Title', 'Category', 'Price'])
print("✅ Extracted Product Data:\n")
print(df)

# -------------------------------------------
# Step 4: Visualize Data using Matplotlib
# -------------------------------------------
plt.figure(figsize=(8, 6))
plt.bar(df['Title'], df['Price'], color='skyblue')
plt.title('Product Price Comparison')
plt.xlabel('Product Title')
plt.ylabel('Price (in $)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
